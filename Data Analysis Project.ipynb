# Data Analysis and Visualization

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)  #printing percentage

plt.figure(figsize = (5,4))
plt.title('Reservation status Count')
plt.bar(['Not canceled','canceled'], df['is_canceled'].value_counts(),edgecolor = 'k', width = 0.7)
plt.show()
#here we checked the cancellation of the bookings.
#value counts represents the category names and values of the column 
```

```python
plt.figure(figsize = (8,4))
axl = sns.countplot(x='hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = axl. get_legend_handles_labels()
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
# (adr)resort hotels are much higher rates that may be the reason but city hotels are having more booking as well as cancellation ration 
# sanitation as well as maintenance should be done more on city hotels to reduce cancellation
```

```python
resort_hotel = df[df['hotel']== 'Resort Hotel'] #low reservation cancel as per the city one.
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel']== 'City Hotel'] # have very high cancellation ratio
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean() # to check the reason through rates
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hortels', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'],label = 'Resort Hotel') # x value index which is date and y = adr
plt.plot(city_hotel.index, city_hotel['adr'],label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
# city hotel price is less then resort on some days it is higher It can be holidays or peak days.
```

```python
# now we will compare on the basis of months (reservation and cancellation)
df['month'] = df['reservation_status_date'].dt.month 
#taking month because we donot have month column 
plt.figure(figsize = (16,8))
axl = sns.countplot(x= 'month', hue='is_canceled', data = df, palette = 'bright')
legend_labels,_ = axl.get_legend_handles_labels()
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not cancelled','cancelled'])
plt.show()
```

```python
 #now we are checking whether adr for months.
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)
data_to_plot = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x='month', y='adr', data=data_to_plot)
plt.show()
```

```python
#comparision on the basis of countries for cancellation.
cancelled_data = df[df['is_canceled'] == 1]
top_10_country =  cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with reservation cancellation', fontsize = 20)
plt.pie(top_10_country, autopct = '%.2f',labels = top_10_country.index)
plt.show()
#this gives an idea that portugal is having maximum cancellation rate that needs to be fixed correctly.
```

```python
df['market_segment'].value_counts(normalize = True) # it means online presence and reservation is higher
```

```python
cancelled_data['market_segment'].value_counts(normalize = True)
# This shows the cancellation. It means that online is both way not better.
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True) 

not_cancelled_data  = df[df['is_canceled']== 0 ]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True) 

plt.figure(figsize =(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label = 'cancelled')
plt.legend()
```
